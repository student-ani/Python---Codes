Draw a line in a diagram from position (1,3) to (2,8) then to (6,1) and finally to position (8,10):

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xaxis = np.array([1,2,6,8])
yaxis = np.array([3,8,1,10])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(xaxis, yaxis)
plt.plot(xaxis, yaxis,'*')
plt.show()
```
