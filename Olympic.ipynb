```python
medal_data = {
    '2020':[{'event': 'Hockey','athlete':'Belgium Men.s Team','medal':'Gold','country': 'Belgium'},
            {'event': 'Athletics', 'athlete': 'Naffisatou Thiam', 'medal': 'Gold', 'country': 'Belgium'},
            {'event': 'cycling', 'athlete': 'wout van Aert', 'medal': 'silver', 'country': 'Belgium'}],
    '2016':[{'event': 'Athletics','athlete':'Naffisatou Thiam','medal':'Gold','country': 'Belgium'},
            {'event': 'Hockey', 'athlete': 'Belgium Men.s Team', 'medal': 'Silver', 'country': 'Belgium'}],
    '2012':[{'event': 'Cycling','athlete':'Gijs Van Hoecke','medal':'Bronze','country': 'Belgium'},
            {'event': 'Judo', 'athlete': 'Charline van Snick', 'medal': 'Bronze', 'country': 'Belgium'}],
    '2008':[{'event': 'Cycling','athlete':'Greg Van Avermaet','medal':'Gold','country': 'Belgium'}],
    '2004':[{'event': 'Tennis','athlete':'Justine Henin','medal':'Gold','country': 'Belgium'}]
  }
def add_details(year,event,athlete,medal,country):
      if year in medal_data:
          medal_data[year].append({'event': event, 'athlete':athlete, 'medal': medal, 'country': country})
      else:
          medal_data[year] = [{'event': event, 'athlete':athlete, 'medal': medal, 'country': country}]
      print(f"Details added for {athlete} in {year}.")

def view_details(year):
    if year in medal_data:
      for record in medal_data[year]:
        print(f"{record['event']}:{record['athlete']} won {record['medal']} for {record['country']}.")
    else:
      print("Data is not available of this Year.")

def view_medal_count_by_country(country):
    count = 0
    for year in medal_data:
        for record in medal_data[year]:
            if record['country'] == country:
                count += 1
    print(f"{country} has won {count} medals in the last 5 years.")    

def search_athlete_performance(athlete):
    found = False
    for year in medal_data:
      for record in medal_data[year]:
        if record['athlete'].lower()==athlete.lower():
          print(f"{athlete} won {record['medal']} in {record['event']} in {year}.")
          found = True
    if not found:
      print(f"Data is not available of this athlete: {athlete}.")

def view_event_details(event):
    found  = False
    for year in medal_data:
      for record in medal_data[year]:
        if record['event'].lower() == event.lower():
          print(f"In {year}, {record['athlete']} won {record['medal']} for {record['country']} in {event}.")
          found = True
    if not found:
      print("Data is not available of this event: {event}.")   

def main():
    while True:
      print("\n-----------------Olympic Data Explorer------------------")   
      print("\nMenu:")    
      print("1. Add Details")
      print("2. View Details by Year")
      print("3. View Medal Count by Country")
      print("4. Search Athlete Performance")
      print("5. View Event Details")
      print("6. Exit")

      choice = input("Enter Your Choice: ")
      if choice == '1':
         year = input("Enter the year: ")
         event = input("Enter the event: ")
         athlete = input("Enter the athlete's name: ")
         medal = input("Enter the medal (Gold/Silver/Bronze): ")
         country = input("Enter the Country name: ")
         add_details(year,event,athlete,medal,country)

      elif choice == '2':
        year = input("Enter the year to view details: ")
        view_details(year)

      elif choice == '3':
        country = input("Enter the Country name to view medal count: ")               
        view_medal_count_by_country(country)

      elif choice == '4':
        athlete = input("Enter the athlete name to view performance: ")
        search_athlete_performance(athlete)

      elif choice == '5':
        event = input("Enter the event to view details: ")
        view_event_details(event)

      elif choice == '6':
        print("Exiting the program.")
        break

      else:
        print("Invalid choice. Please try again.")

if __name__ == "__main__":
     main()
```
