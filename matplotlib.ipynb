```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,3,5,7,11]
plt.plot(x,y,marker='*',mec='red',ms='10')#linestyle='dashed',color='red',linewidth='3')
plt.xlabel('X-axis')              #ms= marker_size
plt.ylabel('Y-axis')
plt.title("Simple Line plot")
plt.show()
```

```python
import numpy as np
y1 = np.array([3,6,2,0,8])
y2 = np.array([1,9,3,5,7])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,3,5,7,11]
plt.scatter(x,y,color='red')
plt.xlabel('Sales')             
plt.ylabel('Price')
plt.title("Simple Scatter plot")
plt.show()
```

```python
products=['Laptop','PC','TV','Refrigerator','Microwave']
sales=[4,7,2,11,5]
plt.bar(products,sales,color='y')  #plt.barh()-for make horizontal
plt.xlabel('products')
plt.ylabel('sales(in 1000)')
plt.title('Simple Bar Plot')
plt.show()
```

```python
#Subplots
x = [1,2,3,4,5]
y1=[7,8,6,10,11]
y2=[1,4,6,8,10]
#create subplots
fig, axs = plt.subplots(2)
axs[0].plot(x,y1)
axs[0].set_title('First Plot')
axs[1].plot(x, y2,'tab:red')
axs[1].set_title('Second Plot')
#Display
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,3,5,7,11]
plt.plot(x,y,color='r')
plt.xlabel('X-axis')              
plt.ylabel('Y-axis')
plt.title("Plot with Annotations")
plt.annotate('Peak', xy=(5,11),xytext=(4,10),arrowprops=dict(facecolor='black',shrink=0.05))
```

```python
#Create line Plot with seaborn
import seaborn as sns
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,3,5,7,11]
sns.lineplot(x=x,y=y)
plt.xlabel('X-axis')             
plt.ylabel('Y-axis')
plt.title("Simple line plot with seaborn")
plt.show()
```
